```python
import os
from os import listdir
from os.path import isfile, join
folder = '/media/larkaa/storage/Pictures/tumblr/'
onlyfiles = []
for root, dirs, files in os.walk(folder): 
        #print(root)
        for filename in files:
            if filename.endswith(('.mp4','mkv','avi')):
                onlyfiles.append(os.path.join(root,filename))
                    
#onlyfiles = [f for f in listdir(folder) if isfile(join(folder, f)) if f.endswith('mp4')]
#print(onlyfiles)
```

```python
#analyze titles
res=[]

for f in onlyfiles:
    cmd = "ffmpeg -i {} -f ffmetadata -hide_banner 2>&1 | grep title".format(f)
    temp= ! eval {cmd}
    
    if any('title' in x for x in temp):
        res.append([f, temp])
```

```python
import pandas as pd
df = pd.DataFrame(res,columns=['file','title'])
temp = [x[0] for x in df['title']]
#print(temp)
df['title'] = temp
#df.title.value_counts()
pd.set_option('display.max_colwidth', -1)
df.sort_values(['title']).groupby('title').filter(lambda x: len(x) > 1)

#df.groupby("sym").filter(lambda x: len(x) == 2)
```
