```python
import requests
import re
import html
import sqlite3
import time
    
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
```

```python
def search_lpsg(search_terms):
    
    s = requests.session()
    
    if type(search_terms) is not list:
        search_terms = [search_terms]
        
    results = []
    
    for terms in search_terms:
        
        data = {'keywords': terms.replace(' ','+')}
        r = s.post('https://www.lpsg.com/search/search',data= data)

        redirect = r.url
        #temp = redirect.split('/?q=')
        #page_url = temp[0] + ('/?page={}&amp;q=') + temp[1]

        if 'LPSG - Error' in r.text or 'No results found' in r.text:
            #print('Not found')
            return None

        results = re.findall('Showing results .* of (\d+)', r.text)[0]

        try:
            max_pages = re.findall('Page . of (\d+)',r.text)[0]
        except:
            max_pages = 1

        res_string = 'LPSG found {} pages {} results: {}'.format(max_pages,results,redirect)
        results.append(res_string)
    
    return(results.join('/n'))

def db_search1(query):
    db_location = '/home/larkaa/code/dn_tools/usr_lists/user_list.db'
    conn = sqlite3.connect(db_location)
    c = conn.cursor()
    c.row_factory = lambda cursor, row: row[0]
    cursor = c.execute(query)
    res = cursor.fetchall()
    conn.close()
    
    return res

def search_mg(search_terms):
    time_sleep=2
    url = 'https://www.malegeneral.com/cam/dongs/'
    chrome_options = Options()  
    chrome_options.add_argument("--headless")
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    ## colab
    #browser = webdriver.Chrome('chromedriver', options=chrome_options)
    browser =  webdriver.Chrome(executable_path='/usr/bin/chromedriver', options=chrome_options)
    # windows
    #browser =  webdriver.Chrome(executable_path='C:\\Users\\Drew\\code\\opera_tools\\chromedriver' , options=chrome_options)

    if type(search_terms) is not list:
        search_terms = [search_terms]
        
    results = []
    
    for terms in search_terms:
        
        browser.get(url)
        
        time.sleep(time_sleep)

        #fill in search
        form1 = browser.find_elements_by_xpath('//input[contains(@class,"gsc-input")]')[0]
        form1.send_keys(terms)

        #press button
        button1 = browser.find_element_by_xpath('//button[contains(@class,"gsc-search-button-v2")]')
        button1.click()
        time.sleep(time_sleep+2)

        #close other tabs
        browser.close()

        #switch to new tab
        browser.switch_to.window(browser.window_handles[0])
        source = browser.page_source

        if 'No Results' in source:
            continue
        else:
            #number of matches
            try:
                num_res = re.findall('About (\d+) result', source)[0]
            except:
                print(terms)
                return(browser.page_source)

            #links on first page
            pages = re.findall('https://www.malegeneral.com.+?php', source)
            pages = list(set(pages))

            results.append([num_res, pages])
            print(terms, num_res, pages)
            #print(pages)
        
        time.sleep(time_sleep)
        
    return(results)
```
